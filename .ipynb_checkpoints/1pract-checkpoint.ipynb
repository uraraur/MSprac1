```python
import pandas as pd
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
```

-------------------------------------------------------------1-------------------------------------------------------------

```python
n = 100
s_mean = 12
var = 2.3**2
a, sigma = 12, 2.3
sample = sts.norm(loc = a, scale = sigma).rvs(size = n)
print(f'Вибіркове середнє: {sample.mean()}, Теоретичне: {s_mean}\nВідхилення: {abs(s_mean - sample.mean())}')
print(f'Вибіркова дисперсія: {sample.var()}, Теоретичне: {var}\nВідхилення: {abs(var - sample.var())}')
plt.hist(sample, 8, color ='c', edgecolor = 'black')
```

```python
n = 300
s_mean = 12
var = 2.3**2
a, sigma = 12, 2.3
sample = sts.norm(loc = a, scale = sigma).rvs(size = n)
print(f'Вибіркове середнє: {sample.mean()}, Теоретичне: {s_mean}\nВідхилення: {abs(s_mean - sample.mean())}')
print(f'Вибіркова дисперсія: {sample.var()}, Теоретичне: {var}\nВідхилення: {abs(var - sample.var())}')
```

```python
def f(a, b, sample):
    x = []
    for i in sample:
        if i >= a and i <= b:
            x.append(i)
    return x
```

```python
low = 9.7
up = 14.3
x = f(low, up, sample)
print(f'Відсоток студентів, які мають частоту дихання в межах від [{low}, {up}]: {100 * len(x)/n} %')
print(f'Відносна частота: {len(x)/n}')
prob = sts.norm(loc = a, scale = sigma).cdf(up) - sts.norm(loc = a, scale = sigma).cdf(low)
print(f'Pr({low} ≤ X ≤ {up}): {prob}')
print(f'Відхилення: {abs(prob - len(x)/n)}')
```



Так як ε1 = ε2 можемо використати нерівність Чебишова:

```python
eps = abs(s_mean - low)**2
print(f'За нерівністю Чебишова: Pr({low} ≤ X ≤ {up}) = Pr(|X - MX| ≤ ε) = 1 - Pr(|X - MX| ≥ ε) ≤ {1 - round(var/eps, 5)}')
```

```python
low = 7.4
up = 16.6
x = f(low, up, sample)
print(f'Відсоток студентів, які мають частоту дихання в межах від [{low}, {up}]: {100 * len(x)/n} %')
print(f'Відносна частота: {len(x)/n}')
prob = sts.norm(loc = a, scale = sigma).cdf(up) - sts.norm(loc = a, scale = sigma).cdf(low)
print(f'Pr({low} ≤ X ≤ {up}): {prob}')
print(f'Відхилення: {abs(prob - len(x)/n)}')
eps = abs(s_mean - low)**2
print(f'За нерівністю Чебишова: Pr({low} ≤ X ≤ {up}) = Pr(|X - MX| ≤ ε) = 1 - Pr(|X - MX| ≥ ε) ≤ {1 - round(var/eps, 5)}')
```

```python
lamb = 0.1
n = 100
sample1 = sts.expon(scale = 1/lamb).rvs(size = n)
print(f'Вибіркове середнє: {sample1.mean()}, Теоретичне: {1/lamb}\nВідхилення: {abs(1/lamb - sample1.mean())}')
print(f'Вибіркова дисперсія: {sample1.var()}, Теоретичне: {1/(lamb**2)}\nВідхилення: {abs(1/(lamb**2) - sample1.var())}')
plt.hist(sample1, 10, color ='c', edgecolor = 'black')
```

Помітимо, що відхилення вибіркової дисперсії доволі велике і кожен раз доволі різне.

```python
low = 8
up = 12
x1 = f(low, up, sample1)
print(f'Частина даних в межах від [{low}, {up}]: {100 * len(x1)/n} %')
prob1 = sts.expon(scale = 1/lamb).cdf(up) - sts.expon(scale = 1/lamb).cdf(low)
print(f'Відносна частота: {len(x1)/n}')
print(f'Pr({low} ≤ X ≤ {up}): {prob1}')
print(f'Відхилення: {abs(prob1 - len(x1)/n)}')
```

```python
x_i = sample1
y_i = sts.norm.ppf(1 - np.exp(-0.1*x_i))
plt.hist(y_i, color ='c', edgecolor = 'black' )  
print(f'Вибіркове середнє: {y_i.mean()}')
print(f'Вибіркова дисперсія: {y_i.var()}')
```

Графік схож на нормальний розподіл, вибіркове середнє ≈ 0, вибіркова дисперсія ≈ 1 	⇒ скоріш за все ця вибірка теж має стандартний нормальний розподіл.
